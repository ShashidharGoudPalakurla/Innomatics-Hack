```python
#importing the required libraries
import pandas as pd
import sqlite3
import json
```

```python
#Loading the CSV Data(orders.csv)
orders_df = pd.read_csv("orders.csv")

print(orders_df.head())
print(orders_df.info())
```

```python
#Loading the JSON Data(users.json)
with open("users.json", "r") as f:
    users_data = json.load(f)

users_df = pd.DataFrame(users_data)

print(users_df.head())
print(users_df.info())
```

```python
#Loading the SQL Data(restaurant.sql)
conn = sqlite3.connect("restaurants.db")
cursor = conn.cursor()

with open("restaurants.sql", "r") as f:
    sql_script = f.read()

cursor.executescript(sql_script)
conn.commit()
```

```python
#Load restaurant table
restaurants_df = pd.read_sql(
    "SELECT * FROM restaurants",
    conn
)

print(restaurants_df.head())
print(restaurants_df.info())
```

```python
#Merge the data (LEFT JOIN)
orders_users_df = pd.merge(
    orders_df,
    users_df,
    on="user_id",
    how="left"
)
```

```python
#Result <--> Restaurants
final_df = pd.merge(
    orders_users_df,
    restaurants_df,
    on="restaurant_id",
    how="left"
)
```

```python
#Creating the final dataset 
final_df.to_csv(
    "final_food_delivery_dataset.csv",
    index=False
)

print("Final dataset created successfully!")
print(final_df.head())
print(final_df.info())
```
